```python
import csv
f=open('cig.csv', 'rb')
reader = csv.reader(f)
tar=[]
nic=[]


for row in reader:
    a = row[0].split()
    b=a[1:3]
    c=map(float,b)
    tar.append(c[0])
    nic.append(c[1])
    

print tar
print nic
    

f.close()
```

```python
from pylab import *
```

```python
scatter(tar, nic)
```

```python
plot(tar, nic)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('axes title')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.scatter(tar, nic)
```
