my dataset that im analyzing is about Melanoma cancer.

```python
import csv
f=open('melanoma.csv', 'rb')
reader= csv.reader(f)
rownum = 0
year=[]
incidence=[]

for row in reader:
    if rownum!=0:
        c = map(float,row[0:])
        year.append(c[0])
        incidence.append(c[1])
    rownum +=1

    
print year
print incidence
      


f.close()



```

```python
from pylab import * 
bar(year,incidence)
```

```python


fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('incidence of melanoma cancer depends on the different year')
ax.set_xlabel('year')
ax.set_ylabel('incidence')
ax.bar(year, incidence)




```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('incidence of melanoma cancer depends on the different year')
ax.set_xlabel('year')
ax.set_ylabel('incidence')
ax.scatter(year, incidence)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('incidence of melanoma cancer depends on the different year')
ax.set_xlabel('year')
ax.set_ylabel('incidence')
ax.plot(year, incidence)
```

the bar,plot and acatter graphs show the connection between the incidence of Melanoma cancer to the increased year. according to my expectetions, the cancer incidence is growing as time goes, but sometimes the incidence of the cancer is decreased as time goes.
